```python
import pandas as pd
```

```python
filename = 'discriminate.csv'
df = pd.read_csv(filename)
df.head()
```

```python
df.iloc[:,:4]
```

```python
# '합계' 컬럼은 모두 숫자.
# '남자', '여자' 컬럼에서는 결측치('-') 존재 ('합계' 컬럼만 쓰면 문제없음)
filename2 = 'foreign_distribution_original.csv'
disDF = pd.read_csv(filename2)
disDF
```

```python
disDF.info()
```

```python
city_names = {
    '서울특별시':'서울', '부산광역시':'부산', '대구광역시':'대구',
    '인천광역시':'인천', '광주광역시':'광주', '대전광역시':'대전',
    '울산광역시':'울산', '세종특별자치시':'세종', '경기도':'경기',
    '강원특별자치도':'강원', '충청북도':'충북', '충청남도':'충남',
    '전라북도':'전북', '전라남도':'전남', '경상북도':'경북',
    '경상남도':'경남', '제주특별자치도':'제주'
}
```

```python
for k, v in city_names.items():
    disDF = disDF.replace(k, v)
```

```python
disDF['행정구역별(1)'].unique()
```

```python
disDF
```

```python
disDF.rename(columns={'행정구역별(1)':'행정구역별', '거주외국인별(1)':'거주외국인별1', '거주외국인별(2)':'거주외국인별2'}, inplace=True)
```

```python
disDF
```

```python
# '행정구역별' 컬럼의 '합계'를 '전국'으로 일괄 변경
disDF = disDF.replace({'행정구역별':'합계'}, '전국')
disDF
```

```python
disDF.to_csv('foreign_distribution.csv', index=False)
```

```python
no_citizen = disDF['거주외국인별(1)'] == '한국국적을 가지지 않은 자 (명)'
yes_citizen = disDF['거주외국인별(1)'] == '한국국적을 취득한 자 (명)'
is_sum = disDF['거주외국인별(2)'] == '소계'
```

```python
no_citizenDF = disDF[no_citizen & is_sum]
no_citizenDF
```

```python
yes_citizenDF = disDF[yes_citizen & is_sum]
yes_citizenDF
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
the_year = disDF['시점'] == 2015
targetDF = disDF[yes_citizen & is_sum & the_year]
```

```python
targetDF['행정구역별(1)'][1:]
```

```python
targetDF['합계'][1:]
```

```python
plt.figure(figsize=(20, 10))
plt.bar(targetDF['행정구역별(1)'][1:], targetDF['합계'][1:])
plt.show()
```

```python
def draw_by_year(cond1, cond2, cond3):
    targetDF = disDF[cond1 & cond2 & cond3]
    plt.figure(figsize=(20, 10))
    plt.bar(targetDF['행정구역별(1)'][1:], targetDF['합계'][1:])
    plt.grid()
    plt.show()
```

```python
draw_by_year(yes_citizen, is_sum, disDF['시점'] == 2015)
```

```python
draw_by_year(yes_citizen, is_sum, disDF['시점'] == 2022)
```

```python
# 다중 그래프
targetDF1 = disDF[yes_citizen & is_sum & (disDF['시점'] == 2015)]
targetDF2 = disDF[yes_citizen & is_sum & (disDF['시점'] == 2022)]
plt.figure(figsize=(15, 10))
plt.title('거주지 별 귀화자 수 (2015년 vs. 2022년)', fontsize=16)
plt.bar(targetDF2['행정구역별(1)'][1:], targetDF2['합계'][1:], color='red')
plt.bar(targetDF1['행정구역별(1)'][1:], targetDF1['합계'][1:])
plt.grid()
plt.show()
```
