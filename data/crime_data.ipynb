```python
import pandas as pd
```

```python
filename = 'crime_by_region_original.csv'
crimeDF = pd.read_csv(filename)
crimeDF
```

```python
is_portion = crimeDF['범죄별(1)'] == 'A/Bⅹ100,000 (건/10만명)'
crimeDF[is_portion]
```

```python
crimeDF['범죄별(2)'].unique()
```

```python
the_year = crimeDF['시점'] == 2014
crimeDF[is_portion & the_year]
```

```python
crimeDF.info()
```

```python
crimeDF.columns = ['범죄별1', '범죄별2', '시점', '발생지역별', '데이터']
```

```python
crimeDF
```

```python
# '발생지역별' 컬럼의 '계'를 '전국'으로 일괄 변경
crimeDF = crimeDF.replace({'발생지역별':'계'}, '전국')
crimeDF
```

```python
# 전처리가 끝난 데이터프레임을 csv 파일로 저장
crimeDF.to_csv('crime_by_region.csv', index=False)
```
