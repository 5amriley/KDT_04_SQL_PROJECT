```python
import pandas as pd
```

```python
filename = 'city_population_original.csv'
popDF = pd.read_csv(filename)
popDF
```

```python
city_names = {
    '서울특별시':'서울', '부산광역시':'부산', '대구광역시':'대구',
    '인천광역시':'인천', '광주광역시':'광주', '대전광역시':'대전',
    '울산광역시':'울산', '세종특별자치시':'세종', '경기도':'경기',
    '강원도':'강원', '충청북도':'충북', '충청남도':'충남',
    '전라북도':'전북', '전라남도':'전남', '경상북도':'경북',
    '경상남도':'경남', '제주특별자치도':'제주'
}
```

```python
for k, v in city_names.items():
    popDF = popDF.replace(k, v)
```

```python
popDF['소재지(시군구)별(1)'].unique()
```

```python
popDF
```

```python
popDF.rename(columns={'소재지(시군구)별(1)':'소재지'}, inplace=True)
```

```python
popDF.to_csv('city_population.csv', index=False)
```
